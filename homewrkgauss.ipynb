```python
import numpy as np
  
def is_pos_def(x):
    return np.all(np.linalg.eigvals(x) > 0)

n = int(input("Enter the size of the matrix: "))
```

```python
is_good_enough = False
while not is_good_enough  :
    x = np.arange(1,n)
    V = []
    for i in range(n):
        y = np.random.randn(n)*2
        V.append(y)
    A = np.cov(V)
    is_good_enough = is_pos_def(A)
A = A
```

```python
%%time

B=np.ones((n,1))

def gauss(A,B,n):
    A = np.hstack((A,B))
    for i in range(n):
        for j in range(i, n):
            if A[j][i] != 0:
                A[i], A[j] = A[j], A[i]
                break
        for j in range(i + 1, n):
            if A[i][i] != 0:
                l = -A[j][i] / A[i][i]
                for k in range(len(A[j])):
                    A[j][k] += l * A[i][k]
    ans = []
    for i in range(n - 1, -1, -1):
        if A[i][i] == 0:
            return None
        
        l = 1 / A[i][i]

        for j in range(len(A[i])):
            A[i][j] *= l
        for j in range(len(ans)):
            for u in range(len(A[i]) - 2 - j, -1, -1): 
                if u != i:
                    A[i][u] *= ans[j]
                    break

        x = A[i][len(A[i]) - 1]
        for j in range(len(A[i]) - 1):
            if j != i:
                x -= A[i][j]
        ans.append(x)
    ans = np.array([ans[el] for el in range(len(ans) - 1, -1, -1)])
    return(ans)
print(gauss(A,B,n))
```
