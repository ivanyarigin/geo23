```python
import numpy as np
  
def is_pos_def(x):
    return np.all(np.linalg.eigvals(x) > 0)

n = int(input("Enter the size of the matrix: "))
```

```python
is_good_enough = False
while not is_good_enough  :
    x = np.arange(1,n)
    V = []
    for i in range(n):
        y = np.random.randn(n)*2
        V.append(y)
    A = np.cov(V)
    is_good_enough = is_pos_def(A)
A = A
```

```python
%%time
def gauss(A,B,n):
    A = np.hstack((A,B))
    n = len(A)
    for i in range(n):
        for j in range(i, n):
            if A[j][i] != 0:
                A[i], A[j] = A[j], A[i]
                break
        for j in range(i + 1, n):
            if A[i][i] != 0:
                l = -A[j][i] / A[i][i]
                for k in range(len(A[j])):
                    A[j][k] += l * A[i][k]
    ans = []
    for i in range(n - 1, -1, -1):
        if A[i][i] == 0:
            return None
        
        l = 1 / A[i][i]

        for j in range(len(A[i])):
            A[i][j] *= l
        for j in range(len(ans)):
            for u in range(len(A[i]) - 2 - j, -1, -1): 
                if u != i:
                    A[i][u] *= ans[j]
                    break

        x = A[i][len(A[i]) - 1]
        for j in range(len(A[i]) - 1):
            if j != i:
                x -= A[i][j]
        ans.append(x)
    ans = np.array([ans[el] for el in range(len(ans) - 1, -1, -1)])
    ans = ans.reshape((n, 1))
    a = np.zeros((n,n))
    for i in range(n):
        for j in range(i, n):
            if i == j:
                a[i][j] = 1
    a = np.hstack((a,ans))
    return(a)

a=A
b=np.ones((n,1))
l = np.zeros((n,n))
l[0][0] = a[0][0]**(0.5)
h = 0
m = 0
for  i in range(1,n):
    for j in range (n):
        if j <= i:
            h = 0
            s = 0
            if i == j:
                for m in range(j):
                    if j == 0:
                        h = 0
                    else:
                        h = h + l[i][m]**2
                l[i][j] = (a[i][j] - h)**(0.5)
            else:
                for m in range(j):
                    if j == 0:
                        s =0
                    else:
                        s = s + l[i][m]*l[j][m]
                l[i][j] = (a[i][j] - s)/l[j][j]
lt=np.transpose(l)
y1 = gauss(l,b,n)
y = np.zeros((n,1))
for k in range(n):
    y[k][0] = y1[k][n]
x1 = gauss(lt,y,n)
x = np.zeros((n,1))
for k in range(n):
    x[k][0] = x1[k][n]
print(x)
```
